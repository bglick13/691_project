```python
import os, sys
os.chdir('..')
os.getcwd()
import numpy as np
import pandas as pd
import pickle
import torch
import matplotlib.pyplot as plt
import seaborn as sns
from torch.utils.data import DataLoader
from models.double_decoder import MovieSequenceEncoder
import torch

%matplotlib inline
```

```python
with open('scripts/history_baseline_50.pickle', 'rb') as f:
    baseline_history = pickle.load(f)

with open('scripts/history_ratings_50.pickle', 'rb') as f:
    ratings_history = pickle.load(f)
    
with open('scripts/history_users_50.pickle', 'rb') as f:
    users_history = pickle.load(f)
    
with open('scripts/history_both_50.pickle', 'rb') as f:
    both_history = pickle.load(f)
```

```python
fig, ax1 = plt.subplots()

color = 'tab:red'
ax1.set_xlabel('step')
ax1.set_ylabel('User Activation', color=color)
ax1.plot(pd.Series(both_history['user_activations'][::100]).rolling(50).mean(), color=color)
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis

color = 'tab:blue'
ax2.set_ylabel('Masked Movie Accuracy', color=color)  # we already handled the x-label with ax1
ax2.plot(pd.Series(both_history['masked_movie_acc']).rolling(50).mean(), color=color)
ax2.tick_params(axis='y', labelcolor=color)

fig.tight_layout()  # otherwise the right y-label is slightly clipped
```

```python
plt.plot(pd.Series(both_history['user_activations'][::100]).rolling(50).mean())
plt.plot(pd.Series(both_history['masked_movie_acc']).rolling(50).mean())
```

```python
user_model = torch.load('scripts/users.torch')
ratings_model = torch.load('scripts/ratings.torch')
```

```python
n_user_p = sum(p.numel() for p in user_model.parameters() if p.requires_grad)
n_ratings_p = sum(p.numel() for p in ratings_model.parameters() if p.requires_grad)
```

```python
n_user_p / n_ratings_p
```

```python
fig, ax = plt.subplots(figsize=(17, 8))
ax.plot(pd.Series(users_history['masked_movie_acc']).rolling(50).mean() / n_user_p)
ax.plot(pd.DataFrame(ratings_history).rolling(50).mean()['masked_movie_acc'] / n_ratings_p)
plt.legend(['Users', 'Ratings'])
plt.title('Train Masked Movie Accuracy per Model Parameter')
plt.ylabel('Accuracy')
plt.savefig('report_7_per_param_acc.png')
```

```python
fig, ax = plt.subplots(figsize=(17, 8))
ax.plot(pd.Series(users_history['masked_movie_acc']).rolling(50).mean().diff(50))
ax.plot(pd.DataFrame(ratings_history).rolling(50).mean().diff(50)['masked_movie_acc'])
plt.legend(['Users', 'Ratings'])
plt.title('Train Masked Movie Accuracy per Model Parameter')
```
